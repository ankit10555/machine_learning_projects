# Zoom effect using xlim ylim

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax1.plot(x,z,color="green")

ax2=fig.add_axes([0.2,0.5,0.4,0.4])
ax2.plot(x,z,color='red')
ax2.set_title("zoom")
ax2.set_xlim(40,60)
ax2.set_ylim(1000,4000)
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,linestyle='--',color='blue',lw=3)
axes[1].plot(x,z,linestyle='-',color='red',lw=3)


plt.show()
```
