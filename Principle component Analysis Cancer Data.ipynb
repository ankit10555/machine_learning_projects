```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(data=cancer['data'],columns=cancer['feature_names'])
```

```python
df
```

```python
cancer['target']
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
model=StandardScaler()
```

```python
df=pd.DataFrame(model.fit_transform(df),columns=df.columns)
```

```python
df
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA(n_components=2)
```

```python
xpca=model.fit_transform(df)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(xpca[:,0],xpca[:,1],c=cancer['target'])
plt.xlabel('first principle label')
plt.ylabel('2nd principle label')
```

```python
model.components_
```

```python
df_comp=pd.DataFrame(model.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
sns.heatmap(df_comp,cmap="plasma")
```
